```python
import matplotlib.pyplot as plt
```

```python
# Count the number of Nationalists
nationality_counts = data['Nationality'].value_counts()
nationality_counts

# Filter Most 20 Nationalists
filtered_nationality_counts = nationality_counts[nationality_counts > 317]

# Combine Nationalists with count <= 5 into 'Others' category
filtered_nationality_counts['Others'] = nationality_counts[nationality_counts <= 317].sum()

plt.figure(figsize=(10, 10))
plt.pie(filtered_nationality_counts.values, labels=filtered_nationality_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Players Most 20 Nationalists ')
plt.show()
```

```python
max_value_player = data.loc[data['Value(in Euro)'].idxmax()] #player with maximum value

max_value_index = data['Value(in Euro)'].idxmax()            #finding a player with a maximum value
player_name = data.loc[max_value_index, 'Full Name']
player_value = data.loc[max_value_index, 'Value(in Euro)']

# Display the player with the maximum value
print("Player with the maximum value:\nName: {}\nValue in Euro: {}".format(player_name, player_value))
```

```python
min_value_player = data.loc[data['Value(in Euro)'].idxmin()] #player with maximum value

min_value_index = data['Value(in Euro)'].idxmin()            #finding a player with a maximum value
player_name = data.loc[min_value_index, 'Full Name']
player_value = data.loc[min_value_index, 'Value(in Euro)']

# Display the player with the maximum value
print("Player with the minimum value:\nName: {}\nValue in Euro: {}".format(player_name, player_value))
```

```python
df_sorted = data.sort_values('Value(in Euro)', ascending=False)  #sorting the data in a descending order

player_name = df_sorted.iloc[:10]['Full Name'] #getting the players with maximum value
player_value = df_sorted.iloc[:10]['Value(in Euro)']

plt.figure(figsize=(10, 6))  #creating a horizontal bar chart
plt.barh(player_name, player_value)
plt.xlabel('Value')
plt.ylabel('Player')
plt.title('Player with the Maximum Value')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Value(in Euro)'])
plt.scatter(df_sorted.iloc[:10]['Age'], df_sorted.iloc[:10]['Value(in Euro)'], color='red', label='Player with Max Value')
plt.xlabel('Age')
plt.ylabel('Value(in Euro)')
plt.title('Player Value vs. Age')
plt.legend()
plt.show()
```

```python
# Sort the dataset by value in descending order
df_sorted = data.sort_values('Overall', ascending=False)

# Get the first players with the maximum value
player_name = df_sorted.iloc[:10]['Full Name']
player_value = df_sorted.iloc[:10]['Overall']

# Create a horizontal bar chart
plt.figure(figsize=(25, 6))
plt.bar(player_name, player_value)
plt.xlabel('Overall')
plt.ylabel('Player')
plt.title('Player with the Maximum Overall ')
plt.show()
```

```python
preferred_foot_count = data["Preferred Foot"].value_counts()
preferred_foot_count
```

```python
plt.figure(figsize=(10, 5))
plt.pie(preferred_foot_count.values, labels=preferred_foot_count.index, autopct='%1.1f%%')
plt.title('Proportion of preferred foot')
plt.show()
```

```python
# Find the first player with the maximum value in each position
max_value_players = data.groupby('Best Position').apply(lambda x: x.loc[x['Value(in Euro)'].idxmax()])

# Create the grouped bar chart
plt.figure(figsize=(20, 6))
positions = max_value_players['Best Position']
values = max_value_players['Value(in Euro)']
players = max_value_players['Known As']
plt.bar(positions, values, label='Players')
plt.xlabel('Position')
plt.ylabel('Value')
plt.title('First Player with Maximum Value in Each Position')
plt.xticks(rotation=45)
plt.legend()

# Annotate player names on the bars
for i, value in enumerate(values):
    plt.text(i, value, players.iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Findings player with the minimum value in each position
min_value_players = data.groupby('Best Position').apply(lambda x: x.loc[x['Value(in Euro)'].idxmin()])

# Create the grouped bar chart
plt.figure(figsize=(20, 6))
positions = min_value_players['Best Position']
values = min_value_players['Value(in Euro)']
players = min_value_players['Known As']
plt.bar(positions, values, label='Players')
plt.xlabel('Position')
plt.ylabel('Value')
plt.title('First Player with Minimum Value in Each Position')
plt.xticks(rotation=45)
plt.legend()

# Annotate player names on the bars
for i, value in enumerate(values):
    plt.text(i, value, players.iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Calculate the total value for each position
total_values = data.groupby('Best Position')['Value(in Euro)'].sum()

# Create the stacked bar chart
positions = max_value_players['Best Position']
player_values = max_value_players['Value(in Euro)']
total_values = total_values[positions]
players = max_value_players['Known As']
plt.figure(figsize=(20, 6))
plt.bar(positions, player_values, label='First Player with Max Value')
plt.bar(positions, total_values - player_values, bottom=player_values, label='Other Players')
plt.xlabel('Position')
plt.ylabel('Value')
plt.title('Comparison of First Player with Max Value in Each Position')
plt.xticks(rotation=45)
plt.legend()

# Annotate player names on the bars
for i, value in enumerate(player_values):
    plt.text(i, value, players.iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Count the number of players in each position
position_counts = data['Positions Played'].value_counts()
position_counts

# Group players based on the number of positions they can play
grouped_players = position_counts.value_counts()
grouped_players

# Filter First 8 positions
filtered_position_counts = position_counts[position_counts > 478]

# Combine positions with count <= 5 into 'Others' category
filtered_position_counts['Others'] = position_counts[position_counts <= 478].sum()

plt.figure(figsize=(10, 10))
plt.pie(filtered_position_counts.values, labels=filtered_position_counts.index, autopct='%1.1f%%')
plt.title('Proportion of First 8 positions')
plt.show()
```
